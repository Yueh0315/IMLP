## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE

```python
#匯入在sklearn.linear_model套件裡面的LinearRegression模型
#匯入在sklearn.model_selection套件裡面的train_test_split模組
#切割數據集(訓練資料80%、測試資料20%，設定random_state=20)
#畫出訓練資料集的matplotlib圖形m

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x = x.reshape(-1, 1)  # 1D to 2D conversion
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=20)
regr_lin = LinearRegression()
regr_lin.fit(x_train, y_train)
plt.scatter(x_train, y_train, color='tab:blue', label='Training Data')
plt.plot(x_train, regr_lin.predict(x_train), color='tab:red', label='Fitted Line')
```

## 將訓練出來的函數預測測試集的X值
1. 使用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape<br>
使用X = 2.44897959，預測出來數值應該為 4.3025375<br>
所有測試集資料
$$\widehat{y}=xw+b=w_{1}x_{1}+b$$

```python
# y_train = w*x_train + b = w1*x_test + b1
# 用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape
# 使用X = 2.44897959，預測出來數值應該為 4.3025375

x = 2.44897959
y = regr_lin.predict(np.array([[x]]))
print(f"預測值: {y[0]}")
```

```python
y_predict = regr_lin.predict(x_test)
w = regr_lin.coef_[0]
b = regr_lin.intercept_
print(f"斜率: {w:.2f}, 截距: {b:.2f}")
```

2. 並計算出測試階段的MSE

```python
# plot target points(blue) and predicted points(red)
# calculate the mean squared error: regr_lin.predict(X)-y)**2) / len(y)
plt.scatter(x_test, y_test, color='tab:blue', label='Ground Truth')
plt.scatter(x_test, y_predict, color='tab:red', label='Predict')
mse = np.sum((y_predict - y_test) ** 2)/ len(y_predict)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'f(x) = 1.6x + 0.2, mse: {mse:.2f}')
plt.legend()
```
